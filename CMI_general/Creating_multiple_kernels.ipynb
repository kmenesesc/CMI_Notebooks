```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
```

```python
#given two vectors
sizeOfVector = 5
#a = 5*np.random.rand(sizeOfVector)
#b = 5*np.random.rand(sizeOfVector)
a = 10*np.random.rand(sizeOfVector)
b = a + .01*10*np.random.rand(sizeOfVector)
```

```python
a = np.array([ 4.71671957 , 2.61179601 , 2.57650199 , 2.81884829 , 0.31051939])
b = np.array([ 4.2776846  , 2.20058435 , 4.49611382 , 0.63765688 , 0.50276141])
aP = np.array([ 4.71671957 , 2.61179601 , 2.57650199 , 2.81884829 , 0.31051939])
bP = np.array([ 4.2776846  , 2.20058435 , 4.49611382 , 0.63765688 , 0.50276141])
```

```python
for i in range (0 , sizeOfVector):
    #print("run number "+ str(i))
    #print("Initially size of a is " + str(a.size) +" and a = ")
    #print(a)
    #create destination for points
    n = 100 #number of points we want to add
    theta = math.pi*np.random.rand(n)
    mu, sigma = 0, 1 #mean and standard deviation
    radii = np.random.normal(mu, sigma, n)
    x = np.zeros(n)
    y = np.zeros(n)
    np.multiply(radii,np.sin(theta),y)
    np.multiply(radii,np.cos(theta),x)
    x = x + a[i]
    y = y + b[i]
    a = np.append(a,x)
    b = np.append(b,y)
    #print("After the size of a is " + str(a.size) + " and a = ")
    #print(a)
    #print("We appended x =")
    #print(x)
    #print("\n")
    
    
```

```python
fig1 = plt.figure()
ax11 = fig1.add_subplot(111)
ax11.scatter(a,b, s = 10 ,c='b', marker="s", label='Recollected Data')
ax11.scatter(aP,bP, s = 10 ,c='r', marker="s", label='Recollected Data')
plt.grid()
plt.show()
```
