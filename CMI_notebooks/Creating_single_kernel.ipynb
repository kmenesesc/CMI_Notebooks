```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
```

First decide the angle of the points

```python
# Try this data point
x_0 = 0
y_0 = 0

#number of points we want to add
n = 1000
```

```python
theta = math.pi*np.random.rand(n)
```

```python
print(theta)
```

Now choose radii data from a normal distribution

```python
mu, sigma = 0, 1 #mean and standard deviation
radii = np.random.normal(mu, sigma, n)
```

```python
print(radii)
```

Now find cartesian coordinates

```python
#print(np.cos(theta))
x = np.zeros(n)
y = np.zeros(n)
np.multiply(radii,np.sin(theta),y)
np.multiply(radii,np.cos(theta),x)
x = x + x_0
y = y + y_0
```

```python
fig1 = plt.figure()
ax11 = fig1.add_subplot(111)


ax11.scatter(x,y, s = 10, c='r', marker="o", label='Original Data')
ax11.scatter(x_0,y_0, s = 10 ,c='b', marker="s", label='Recollected Data')
#plt.legend(loc='upper left');
plt.grid()
plt.show()
```
